```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import accuracy_score
datos = {
    'pabras_spam': [3,7,1,8],
    'longitud_mensaje': [300,200,100,50],
    'es_spam': [1,0,1,0]
}

df = pd.DataFrame(datos)

x = df[['pabras_spam', 'longitud_mensaje']]
y = df['es_spam']

print(x.shape)
print(y.shape)
```

Redes neuronales

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
# Define los datos de entrada (features) como un array de NumPy. Cada fila es una muestra, cada columna una característica.
x = np.array([[0,0],[0,1],[1,0],[1,1]])
# Define las etiquetas de salida (target) como un array de NumPy. Corresponden a las etiquetas de cada muestra en 'x'.
y = np.array([0,1,1,0])

# Divide los datos en conjuntos de entrenamiento y prueba.
# x_train: datos de entrada para entrenamiento
# x_test: datos de entrada para prueba
# y_train: etiquetas de salida para entrenamiento
# y_test: etiquetas de salida para prueba
# test_size=0.25: el 25% de los datos se usarán para prueba
# random_state=42: asegura que la división sea la misma cada vez que se ejecute el código
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state= 42)

# Inicializa un modelo de red neuronal secuencial. Las capas se añadirán una tras otra.
modelo = Sequential()

# Añade una capa densa (totalmente conectada) al modelo.
# 4: número de neuronas en esta capa
# input_dim=2: el número de características de entrada (2 en este caso: 0s y 1s)
# activation='relu': función de activación ReLU (Rectified Linear Unit)
modelo.add(Dense(4, input_dim=2, activation='relu'))
# Añade la capa de salida.
# 1: número de neuronas en la capa de salida (para clasificación binaria)
# activation='sigmoid': función de activación sigmoide, que produce una salida entre 0 y 1 (probabilidad)
modelo.add(Dense(1, activation='sigmoid'))

# Compila el modelo. Esto configura el proceso de entrenamiento.
# loss='binary_crossentropy': función de pérdida para clasificación binaria
# optimizer='adam': algoritmo de optimización Adam
# metrics=['accuracy']: métrica para evaluar el rendimiento (precisión)
modelo.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entrena el modelo con los datos de entrenamiento.
# x_train, y_train: datos de entrenamiento y sus etiquetas
# epochs=100: número de veces que el modelo pasará por todo el conjunto de entrenamiento
# batch_size=1: número de muestras por actualización de gradiente
# verbose=0: no muestra el progreso del entrenamiento durante la ejecución
modelo.fit(x_train, y_train, epochs=100, batch_size=1, verbose=0)

# Evalúa el modelo con los datos de prueba.
# _, accuracy: la función devuelve la pérdida y la precisión; solo necesitamos la precisión (el _ descarta la pérdida)
_, accuracy = modelo.evaluate(x_test, y_test)

# Imprime la precisión del modelo en los datos de prueba.
# %.2f % (accuracy*100): formatea la precisión como un porcentaje con dos decimales
print('Accuracy: %.2f' % (accuracy*100))

# Realiza predicciones en los datos de prueba.
resultados = modelo.predict(x_test)
# Imprime las predicciones redondeadas a dos decimales.
# f'probabilidad de estar enfermo': texto descriptivo
# resultados.round(2): redondea las predicciones a 2 decimales
print(f'probabilidad de estar enfermo',resultados.round(2))
```
