```python
from datasets import load_dataset, DatasetBuilder
import shutil
import os
```

```python
dataset_cache_dir = os.path.expanduser("~/.cache/huggingface/datasets/ikuldeep1")
if os.path.exists(dataset_cache_dir):
    shutil.rmtree(dataset_cache_dir)

ds = load_dataset("ikuldeep1/vehicle-damage-fraud-image")
```

```python
print(ds)  # Información general
print(ds['train'][0])  # Primer ejemplo
ds['train'].features  # Estructura del dataset
```

```python
import matplotlib.pyplot as plt

# Mostrar 5 ejemplos con sus etiquetas
for i in range(10):
    example = ds['train'][i]
    plt.imshow(example['image'])
    plt.title(f"Label: {example['label']}")
    plt.axis("off")
    plt.show()
```
